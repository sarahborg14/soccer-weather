```python
#may need this PDF splitter function if wanting to work with individual player stats
    #which are on the second page of each pdf 

def pdf_splitter(path):
    fname = os.path.splitext(os.path.basename(path))[0]
    
    pdf = PdfFileReader(path)
    for page in range(pdf.getNumPages()):
        pdf_writer = PdfFileWriter()
        pdf_writer.addPage(pdf.getPage(page))
        output_filename = '{}_page_{}.pdf'.format(
            fname, page+1)
        
        with open(output_filename, 'wb') as out:
            pdf_writer.write(out)
            
        print('Created: {}'.format(output_filename))
```

```python
pdf_splitter("SoccerWomen2013.pdf")
```

```python
##Still a bug here...

df = {}
for n in enumerate(file_list):
    print(n, n[1]) #this returns a tuple with an index and the associated value
    m = n[1][0:15]
    df[m] = tabula.read_pdf('soccerstats/' +n[1])
    first_line = list(df[m].head(0))
    print(first_line)
    data = []
    if len(first_line) < 8: 
        df[m].columns = ["date", "opponent", "w/l", "score", "ot", "attend", "site"]
        data.insert(0, {"date": first_line[0], "opponent":first_line[1], "w/l":first_line[2], "score": first_line[3],
                 "ot": first_line[4], "attend": first_line[5], "site": first_line[6]})
    else: 
        df[m].columns = ["date", "opponent", "w/l", "score", "ot", "attend", "site", "notes"]
        data.insert(0, {"date": first_line[0], "opponent":first_line[1], "w/l":first_line[2], "score": first_line[3],
                 "ot": first_line[4], "attend": first_line[5], "site": first_line[6], "notes": first_line[7]})
    
    df[m] = pd.concat([pd.DataFrame(data), df[m]], ignore_index=True, sort = False)
    
    if n[0] == 0: 
        print(n[0])
        df_full = df[m]
    
    elif n[0] > 0: 
        print(n[0])
        df_full = pd.concat([df_full, df[m]], ignore_index=True, sort = False)

    
print(df_full)
```
